```python
df = pd.read_csv('house_data.csv')
df.head()
```

### 1.2 Изучите стоимости недвижимости

```python
plt.hist(df['price'])
plt.title('Стоимость недвижемости')
plt.xlabel('Цена')
plt.ylabel('Кол-во');
```

### 1.3 Изучите распределение квадратуры жилой

## Условие 2: 

### 2 задача
### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график<br>
Сделайте выводы


```python
data = df['waterfront'].value_counts()
plt.pie(data.values, labels=data.index, autopct='%1.1f%%')
plt.title('Вид на набережную')
plt.legend(['Нету', 'Есть']);
```

### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
plt.bar(data.index, data.values)
plt.title('Распределение этажей домов')
plt.xticks(rotation=50);
```

### 2.3 Изучите распределение состояния домов

```python
plt.hist(df['condition'], bins=30)
plt.title('Распределение состояния домов')
plt.xlabel('Состояния')
plt.ylabel('Кол-во');
```

## Условие 3:

### 3 задача
### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.figure(figsize=(8, 6))
sns.boxenplot(y=df['price'], x=df['grade'], showfliers=False)
plt.title('Взаимосвязь стоимости дома и класса дома')
plt.xlabel('Класс')
plt.ylabel('Цена');
```

Можно сделать вывод, что чем выше класс дома, тем выше цена.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
plt.title('Взаимосвязь стоимости дома и жилой площади')
plt.xlabel('Цена')
plt.ylabel('Жилая площадь');
```

Можно сделать вывод, что чем больше жилая площадь, тем более высока цена.

```python
plt.scatter(df['price'], df['yr_built'])
plt.title('Взаимосвязь стоимости дома и года постройки')
plt.xlabel('Цена')
plt.ylabel('Год');
```

Сильной зависимости стоимости дома от года постройики, не выявлена.

```python
plt.scatter(df['price'], df['sqft_lot'])
plt.title('Взаимосвязь стоимости дома и площади участка')
plt.xlabel('Цена')
plt.ylabel('Площадь участка');
```

Зависимости стоимости дома от размера площади участка, не выявлена.
