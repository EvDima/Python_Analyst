## Условие 2:
На складе лежат разные фрукты в разном количестве.<br>
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

```python
def sums(data: dict) -> int:
    sum = 0
    for i in data:
        sum += data[i]
    return sum

fruits = {
    "Яблоки": 100, 
    "Апельсины": 200,
    "Бананы": 400, 
    "Арбузы": 150, 
    "Нектарины": 100,
    "Ананасы": 50 
}

print(f"Общее количество фруктов на складе: {sums(fruits)}")
```

### Условие 4.2 
Найдите выручку компании в зависимости от месяца<br>
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.<br>
Используйте аннотирование типов.

```python
def revenue_m(date: list, revenue: list) -> dict:
    date = [(i[0][5:7], i[1]) for i in zip(date, revenue)]
    rev = {i[0]:0 for i in date}
    for i in date:
        rev[i[0]] += i[1]
    return rev

revenue_m(date, revenue)
```
